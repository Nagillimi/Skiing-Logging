## Optimal Baseline Index Testing

Concluding on the optimal idx for the baseline roll, the best will ideally have **minimal std between carving angles**. This is a big assumption and may require further investigation into those portions of the runs where a consistent cadence was observed (and thus a consistent carving angle). For now, consider the entire downhill track. The following methods will be compared:

1. $argmin(F(t))$
2. $argmin(\frac{dF(t)}{dt})$
3. $0.5*[argmin(F(t)) + argmin(\frac{dF(t)}{dt})]$
4. $0.5*[argmax(roll(t)) + argmin(roll(t-n))]$
5. $argmean(roll(t-n))$

More of a gut check to identify which baseline is closest and produces the least number of outliers. Method 4&5 will not be used, which present median comparisons on the timing between the peak-peak's.

```python
from domain.sessions import Session_2023_12_30, Session_2023_12_31

session_2023_12_30 = Session_2023_12_30()
```

```python
from utilities.sig_proc import std

carves_1 = [turn.carving_angle_1 for turn in session_2023_12_30.tile.turns]
carves_2 = [turn.carving_angle_2 for turn in session_2023_12_30.tile.turns]
carves_3 = [turn.carving_angle_3 for turn in session_2023_12_30.tile.turns]
carves_4 = [turn.carving_angle_4 for turn in session_2023_12_30.tile.turns]
carves_5 = [turn.carving_angle_5 for turn in session_2023_12_30.tile.turns]

print('std carves_1:', std(carves_1))
print('std carves_2:', std(carves_2))
print('std carves_3:', std(carves_3))
print('std carves_4:', std(carves_4))
print('std carves_5:', std(carves_5))
```

```python
from matplotlib import pyplot as plt
import numpy as np

def stdBarPlot(x):
    _ = plt.figure(figsize = (10, 3))
    plt.bar(np.arange(len(x)), x)
    plt.show()
```

```python
stdBarPlot(carves_1)
stdBarPlot(carves_2)
stdBarPlot(carves_3)
stdBarPlot(carves_4)
stdBarPlot(carves_5)
```
