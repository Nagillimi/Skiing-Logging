## Optimal Baseline Index Testing

Concluding on the optimal idx for the baseline roll, the best will ideally have **minimal std between carving angles**. This is a big assumption and may require further investigation into those portions of the runs where a consistent cadence was observed (and thus a consistent carving angle). For now, consider the entire downhill track. The following methods will be compared:

1. $argmin(F(t))$
2. $argmin(\frac{dF(t)}{dt})$
3. $0.5*[argmin(F(t)) + argmin(\frac{dF(t)}{dt})]$
4. $0.5*[argmax(roll(t)) + argmin(roll(t-n))]$
5. $argmean(roll(t-n))$

More of a gut check to identify which baseline is closest and produces the least number of outliers. Method 4&5 will not be used, which present median comparisons on the timing between the peak-peak's.

```python
from matplotlib import pyplot as plt
import numpy as np


def stdBarPlot(x):
    _ = plt.figure(figsize = (20, 2))
    plt.bar(np.arange(len(x)), x, 1)
    plt.show()
```

```python
import statistics
from models.tile import Tile


def compareBaselines(tile: Tile):
    print('Number of turns:', len(tile.turns))

    carves_1 = [turn.carving_angle_1 for turn in tile.turns]
    carves_2 = [turn.carving_angle_2 for turn in tile.turns]
    carves_3 = [turn.carving_angle_3 for turn in tile.turns]
    carves_4 = [turn.carving_angle_4 for turn in tile.turns]
    carves_5 = [turn.carving_angle_5 for turn in tile.turns]

    print('\tstdev carves_1:', statistics.stdev(carves_1), '\tmean carves_1:', statistics.mean(carves_1), '\tmedian carves_1:', statistics.median(carves_1))
    print('\tstdev carves_2:', statistics.stdev(carves_2), '\tmean carves_2:', statistics.mean(carves_2), '\tmedian carves_2:', statistics.median(carves_2))
    print('\tstdev carves_3:', statistics.stdev(carves_3), '\tmean carves_3:', statistics.mean(carves_3), '\tmedian carves_3:', statistics.median(carves_3))
    print('\tstdev carves_4:', statistics.stdev(carves_4), '\tmean carves_4:', statistics.mean(carves_4), '\tmedian carves_4:', statistics.median(carves_4))
    print('\tstdev carves_5:', statistics.stdev(carves_5), '\tmean carves_5:', statistics.mean(carves_5), '\tmedian carves_5:', statistics.median(carves_5))

    stdBarPlot(carves_1)
    stdBarPlot(carves_2)
    stdBarPlot(carves_3)
    stdBarPlot(carves_4)
    stdBarPlot(carves_5)
```

```python
compareBaselines(session_2022_12_27.tile)
```

```python
compareBaselines(session_2023_12_30.tile)
```

```python
compareBaselines(session_2024_01_01.tile)
```
