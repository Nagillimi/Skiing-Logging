## Hypotheses

- aligned euler data will provide accurate representations for edge angle, slope grade, and heading (r, p, y)
- it's possible to isolate the ski angle of attack based on the direction of turning tangential velocites wrt to heading

## Elements 

- track differences in tangential angles to ski angle (angle of attack)
- carving/skidding estimation
- minimum turning radius


```python
from domain.sessions import Session_2023_12_30

session_2023_12_30 = Session_2023_12_30()
```

```python
from matplotlib import pyplot as plt
import numpy as np

def plotLine(x):
    plt.plot(np.arange(len(x)), x)
```

```python
from constants.ski_th import SKI_SIDECUT_R
from models.turn import Turn

def plotCarvingAngleAnalysis(turn: Turn):
    offset_abs_roll = turn.offset_abs_roll
    turning_radii = SKI_SIDECUT_R * np.cos(np.deg2rad(offset_abs_roll))

    fig, ax = plt.subplots(2, 1)
    ax[0].plot(np.arange(len(offset_abs_roll)), offset_abs_roll)
    ax[0].set_title('Ski Carving Angle (Roll) [deg]')

    ax[1].plot(np.arange(len(turning_radii)), turning_radii)
    ax[1].set_title('Ski Turning Radius [m]')

    plt.tight_layout()
    plt.show()
```

```python
[plotCarvingAngleAnalysis(turn) for i, turn in enumerate(session_2023_12_30.tile.turns) if i < 10]
```
